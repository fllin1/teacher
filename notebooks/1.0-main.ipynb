```python
import os
from docx import Document
```

```python
def get_all_files(folder_path):
    return [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and f.endswith('.docx')]

def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return ' '.join(full_text)
```

```python
folder_path = '../data/raw'
file_paths = sorted(get_all_files(folder_path))
lessons = {}
vocab = {}

for file_path in file_paths:
    lessons[file_path[:10]] = read_docx(os.path.join(folder_path, file_path)).split('  ')    
    
for lesson in lessons:
    for word in lessons[lesson]:
        word = ''.join(e for e in word if e.isalnum())
        if word.startswith(' '):
            word = word[1:]
        if word not in vocab and 4 >= len(word) >= 1:
            vocab[word] = {}
```

```python
import requests
from bs4 import BeautifulSoup

def search_mandarin_dictionary_requests(query):
    url = "https://chine.in/mandarin/dictionnaire/"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
    }
    data = {
        'q': query,
        'Submit': '1'
    }

    response = requests.post(url, headers=headers, data=data)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")
    results = soup.find_all(class_="table invert_img")

    for result in results:
        print(result.get_text())

search_mandarin_dictionary_requests("你好")
```
